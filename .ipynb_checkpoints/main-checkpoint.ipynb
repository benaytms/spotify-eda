## **Exploratory Data Analysis (EDA)**
It's the process of understanding and summarizing data sets, often using visual plots and analyzing the shape and depth of it.

### **Data set**

In this example i'll be using the "spotify tracks dataset" uploaded by Maharshipandya on Kaggle.  

I'll perform some basic EDA methods and try gather some conclusion by the end of this notebook.

```python
import kagglehub   # Kaggle package for downloading their data sets

path = kagglehub.dataset_download("maharshipandya/-spotify-tracks-dataset")

print("path to dataset: ", path)
```

### 1. We import the required libraries

For EDA processes the following are standards and almost everytime you'll be using them for Data Science/Analysis projects.

***Numpy*** is a library used in scientific computing. It adds support for large arrays and matrices with good optimization, and many mathematical functions also.

***Pandas*** complement *Numpy* very well, being a library for data manipulation and analysis - many of its components actually use Numpy itself. It offers data structures operations with support for numerical tables and time series.

***Matplotlib*** as the name suggest is a library for numerical and plotting purposes. Contains an object-oriented API for embedding plots into applications using general purpose GUI toolkits like Tkinter, Qt, GTK, etc. In our case, we'll only need the **pyplot** method.

***seaborn*** is a data visualization library based on matplotlib. It provides high-level interface for drawing attractive and informative statistical graphics.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
## the above line guarantees the matplotlib plot will be inside the cell
sns.set(color_codes=True)
```

*Series* objects being the array of Numpy. So a DataFrame can be summarized basically as a 2D Series. See below.

```python
df = pd.read_csv(path + "/dataset.csv")
# the df name is commonly used to represent the DataFrame being worked on

df.head(5)
# the head(x) function shows the first x values
```

```python
df.tail(5)
# the tail(x) function shows the last x values
```

### 3. Backup Original DataFrame
Before we go any further, one of the most important steps is to work on a copy of the DataFrame, and leave the original DataFrame unchanged - if we need that specific DataFrame to be the one with changes we can make at the end after making sure everything is in order with the Copied DataFrame

```python
df_copy = df.copy(deep=True)
# the deep parameter makes sure any changes on df_copy won't affect the original df
```

### 4. Data Types
Now we check for irregularities in datatypes, in case any of them seem weird we change to something that makes more sense. In this case, everything seems fine.

```python
df_copy.dtypes
```

### 5. Removing unnecessary columns
At first sight i already saw an unnecessary collumn, 'Unnamed: 0', which is just the DataFrame index. There's also the song characteristics like 'danceability', 'energy', 'loudness', etc. I consider these irrelevant for a simple EDA, so we'll get rid of them as well.

```python
df_copy = df_copy.drop(['Unnamed: 0', 'track_id', 'danceability', 'energy', 
                        'loudness', 'mode', 'speechiness', 'acousticness', 
                        'instrumentalness', 'liveness', 'valence'], axis=1)
# axis=1 specifies that are the columns we are dropping

print("previous number of columns: ", df.shape[1], 
      "\nnew number of columns: ", df_copy.shape[1])
```

```python
df_copy.dtypes
```

### 6. Better names
We can also change the columns names in order to get a more simplified and neat visualization. The most important to change is duration_ms, being that i'll change from miliseconds to just seconds.

```python
df_copy = df_copy.rename(columns={"album_name": "album", 
                                 "track_name": "track",
                                 "duration_ms": "duration",
                                 "time_signature": "t_signature",
                                 "track_genre": "genre"})
df_copy.head(10)
```

### 7. Drop duplicates

Another useful thing to do in a large data set is assecure there aren't no duplicate entries. So first we check how many rows we have, then we check how many duplicates. Then we drop those, and compare the values.

```python
print("total rows: ", df_copy.shape[0])

df_copy_duplicates = df_copy[df_copy.duplicated()]
print("duplicate rows: ", df_copy_duplicates.shape[0])

# 577 out of 114000 are duplicates, which means
print("{:.2f}% of rows are duplicates"
      .format((df_copy_duplicates.shape[0]/df_copy.shape[0])*100))
```

We can take a better look at the count of rows using the .count() method

```python
df_copy.count()
```

In this case, it didn't tell us much besides there are some entries without the artist name, album, or track - We'll check on this after dropping. Now let's drop the 577 duplicates.

```python
df_copy = df_copy.drop_duplicates()
df_copy.count()
```

Now as we can see the values were subtracted in 577.

### 8. Missing or Null values
This is similar to the previous step but before just dropping all the missing/null values, we'll check on them to see what exactly is going on.

For this, we use the methods isnull() and sum() - isnull() detects missing values, sum() sums all of these detected missing values.

```python
df_copy.isnull().sum()
```

As we saw ealier, only these 3 columns have missing value and it's only one - so let's see if it's the same.

We can use the loc[] method for this - loc[] finds values based on a label.  
Basically what it does is that it searches in df_copy, entries that match what's inside the brackets - in this case, searches for **any** missing value in the row axis.

And as we can see, there's only this one occasion - It's a k-pop song, there's not much info on it, but something we can try to do is to go back on the original DataFrame, see if there's **track_id** and see what song is it on Spotify. And then maybe fill in the gaps.

```python
df_copy.loc[df_copy.isnull().any(axis=1)]
```

```python
df.loc[df.isnull().any(axis=1)]
```

After searching on Spotify using the track_id it wasn't able to find what song this was, so we don't have much choice besides dropping it from the DataFrame.

```python
df_copy = df_copy.dropna()
df_copy.count()
```

### Fixing the 'duration' column
As said before, the duration of the tracks are in miliseconds - and this is very unusual for a variable, so let's change it to seconds instead, which is the international standard.

For that, all we need is to divide all entries for the column 'duration' by 1000.

```python
df_copy['duration'] = df_copy['duration']/1000
```

```python
df_copy.head(5)
```

Now the duration is in seconds, but we don't really need to be so specific with decimals, so we'll make the dtype int64.

Something only now i noticed is that the *time signature* is just one number - and as you may know time signature is the number of beats in a measure of time, so having just one number doesn't give us anything really. Also, the tempo is in decimals, which is a little peculiar.

So with that assessed, i'll remove the time signature column and change the types for the duration and tempo from float to integer.

```python
df_copy = df_copy.drop('t_signature', axis=1)
df_copy['duration'] = df_copy['duration'].astype('int64')
df_copy['tempo'] = df_copy['tempo'].astype('int64')
```

I'll drop the *key* column as well.

```python
df_copy = df_copy.drop('key', axis=1)
```

```python
df_copy.head(5)
```

Now everything is much more clear and neat.

### 9. Detecting Outliers
Outliers are data points in which lies at an abnormal distance from other points in a random sample. Best way to examplify what an outlier is, is by using the player score analogy - say you are analyzing football players scores for this season, and the average score is around 1000, but you notice few players have a score less than 50 - Those are outlier points.

Outliers can be classified in 3 types, *Global outliers*, *Contextual outliers*, and *Collective outliers*. I won't get much into detail on what each of these mean but in case you want to know more about it you can use this [link](https://www.anodot.com/blog/quick-guide-different-types-outliers/).

In this EDA we won't use statistical methods of detecting outliers, instead we'll focus more on the visualization methods (Box plot, Scatterplot).

Let's start by analyzing the popularity column

```python
sns.boxplot(data=df_copy, x='popularity')
plt.show()
```

The popularity is most concentrated around 20-50 popularity points.

Let's use the describe method to compare.

```python
df_copy['popularity'].describe()
```

The popularity doesn't seem to have much outliers.

We can also test the boxplot on the duration column.

```python
sns.boxplot(data=df_copy, x='duration')
plt.show()
```

```python
df_copy['duration'].describe()
```

Most tracks seem to be around 227 seconds, or approximately 3.5 seconds

But there are plenty of outliers with more than 1000 seconds - which is 16 minutes.  
Which could be some more progressive inclined songs. 

Considering longer songs tend to be less popular, let's see if this makes sense using a **Scatterplot**

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.scatter(df_copy['duration'], df_copy['popularity'])
ax.set_xlabel('duration')
ax.set_ylabel('popularity')
plt.show()
```

As we predicted, longer songs seem to not surpass 60 of popularity.

There's one track with more than 5000 seconds! Which is around 1h20min.  
Let's see the 5 longest songs we have in here.

```python
df_copy.nlargest(5, 'duration')
```

So seems that song with 1h20min is 'Unity(Voyage Mix) Pt. 1' and its popularity is 35 which is unexpected to say the least.

We can also check how popular artists are using **histograms**.

The process we use involves the groupby() method, which basically groups columns based on other methods applied. It's easier if showed.

```python
artist_popularity = df_copy.groupby('artists')['popularity'].sum()
```

Now the *artist_popularity* variable has all artists popularity summed up. We can use the nlargest() method again to get the 30 most popular artists.

```python
top_30 = artist_popularity.nlargest(30)
plt.figure(figsize=(10, 6))

plt.bar(top_30.index, top_30.values, color='blue', edgecolor='black')

plt.xticks(rotation=90)
plt.xlabel('Artists')
plt.ylabel('Total Popularity')
plt.title('Top 30 Most Popular Artists')

plt.show()
```
